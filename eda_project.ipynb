#### **6. Outlier Detection**

```python
column_name = df.select_dtypes(include= 'number').columns

for col in column_name:
    q1 = df[col].quantile(0.25) 
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = df[(df[col] < lower) | (df[col] > upper)][col]

    print("\nColumn :", col)
    print(f"Total Outliers : {len(outliers)}")
    print("Outlier Values :", outliers.values)
```

#### **7. Conditional Probabilities**

##### **Product – Gender**
- Percentage of a Male customer purchasing a treadmill 
- Percentage of a Female customer purchasing KP781 treadmill 
- Probability of a customer being a Female given that Product is KP281

```python
frequency_table_gender = pd.crosstab(df['Product'], df['Gender'])
frequency_table_gender
```

```python
# Percentage of a Male customer purchasing a treadmill 

total_customers = df['Gender'].count()
print("Total Customers =", total_customers)

male_customer_count = (df['Gender'] == 'Male').sum()
print("Male Count =", male_customer_count)

male_customer_pct = (male_customer_count / total_customers) * 100
print(f"Percentage of a Male customer purchasing a treadmill = {male_customer_pct:.2f}%")
```

```python
# Percentage of a Female customer purchasing KP781 treadmill

total_females = df[df['Gender'] == 'Female']['Gender'].count()
print(total_females)

female_kp781 = df[(df['Gender'] == 'Female') & (df['Product'] == 'KP781')]['Gender'].count()
print("Count of Female customer who purchased KP781 =", female_kp781)

female_kp781_pct = (female_kp781 / total_females) * 100
print(f"Percentage of a Female customer purchasing KP781 = {female_kp781_pct:.2f}%")
```

```python
# Probability of a customer being a Female given that Product is KP281

count_kp281 = (df['Product'] == 'KP281').sum()
print("Count of KP281 =", count_kp281)

female_kp281 = df[(df['Gender'] == 'Female') & (df['Product'] == 'KP281')]['Gender'].count()
print("Count of Female who purchased KP281 =", female_kp281)

female_customer_given_kp281 = (female_kp281 / count_kp281)
print(f"Probability of a customer being a Female given that Product is KP281 = {female_customer_given_kp281:.2f}")
```

**Insights**
- Percentage of a Male customer purchasing a treadmill = **57.78%**
- Percentage of a Female customer purchasing KP781 = **9.21%**
- Probability of a customer being a Female given that Product is KP281 = **0.50**
